#Computing Probabilities - Marginal and Conditional Probability

### Marginal Probability



```python
df['Product'].value_counts(normalize=True).mul(100).round(2)
```

### Conditional Probability


```python
pd.crosstab(df['Gender'],[df['Product']], normalize=True, margins=True, margins_name='Total').mul(100).round(2)
```

```python
pd.crosstab(index=df["MaritalStatus"],columns=df["Product"],margins=True,normalize=True).mul(100).round(2)
```

### Probability of each product for given gender

```python
KP281	 = df[df["Product"]=="KP281"]
KP481	 = df[df["Product"]=="KP481"]
KP781	 = df[df["Product"]=="KP781"]

KP281_Male= (len(KP281[KP281["Gender"]=="Male"])/len(df[df["Gender"]=="Male"])*100)
KP481_Male= (len(KP481[KP481["Gender"]=="Male"])/len(df[df["Gender"]=="Male"])*100)
KP781_Male= (len(KP781[KP781["Gender"]=="Male"])/len(df[df["Gender"]=="Male"])*100)

KP281_Female= (len(KP281[KP281["Gender"]=="Female"])/len(df[df["Gender"]=="Female"])*100)
KP481_Female= (len(KP481[KP481["Gender"]=="Female"])/len(df[df["Gender"]=="Female"])*100)
KP781_Female= (len(KP781[KP781["Gender"]=="Female"])/len(df[df["Gender"]=="Female"])*100)

print(f"P(KP281|Male): {KP281_Male:.2f}")
print(f"P(KP481|Male): {KP481_Male:.2f}")
print(f"P(KP781|Male): {KP781_Male:.2f}")

print("\n")

print(f"P(KP281|Female): {KP281_Female:.2f}")
print(f"P(KP481|Female): {KP481_Female:.2f}")
print(f"P(KP781|Female): {KP781_Female:.2f}")
```

### Probability of each product for given Marital Status

```python
KP281	 = df[df["Product"]=="KP281"]
KP481	 = df[df["Product"]=="KP481"]
KP781	 = df[df["Product"]=="KP781"]

KP281_Partnered= (len(KP281[KP281["MaritalStatus"]=="Partnered"])/len(df[df["MaritalStatus"]=="Partnered"])*100)
KP481_Partnered= (len(KP481[KP481["MaritalStatus"]=="Partnered"])/len(df[df["MaritalStatus"]=="Partnered"])*100)
KP781_Partnered= (len(KP781[KP781["MaritalStatus"]=="Partnered"])/len(df[df["MaritalStatus"]=="Partnered"])*100)

KP281_Single= (len(KP281[KP281["MaritalStatus"]=="Single"])/len(df[df["MaritalStatus"]=="Single"])*100)
KP481_Single= (len(KP481[KP481["MaritalStatus"]=="Single"])/len(df[df["MaritalStatus"]=="Single"])*100)
KP781_Single= (len(KP781[KP781["MaritalStatus"]=="Single"])/len(df[df["MaritalStatus"]=="Single"])*100)

print(f"P(KP281|Partnered): {KP281_Partnered:.2f}")
print(f"P(KP481|Partnered): {KP481_Partnered:.2f}")
print(f"P(KP781|Partnered): {KP781_Partnered:.2f}")

print("\n")

print(f"P(KP281|Single): {KP281_Single:.2f}")
print(f"P(KP481|Single): {KP481_Single:.2f}")
print(f"P(KP781|Single): {KP781_Single:.2f}")
```
